```
def preprocess_input(images):
  img = tf.cast(images,'float32')
  img = tf.keras.applications.resnet50.preprocess_input(img)
  return img
```

```
def predefine_model(inputs):
  resnet = tf.keras.applications.resnet.ResNet50(weights='imagenet',include_top=False,input_shape=(224,224,3))
  x = resnet(inputs)
  return x 
```

```
def base_model(inputs):
  x = predefine_model(inputs)
  x = tf.keras.layers.GlobalAveragePooling2D()(x)
  x= tf.keras.layers.Dense(1024,activation='relu')(x)
  x  = tf.keras.layers.Dense(512,activation='relu')(x)
  x = tf.keras.layers.Dense(10,activation ='softmax',name='classification')(x)
  return x 
```

```
def final_model():
  inputs = tf.keras.layers.Input(shape=(32,32,3))
  up_sample = tf.keras.layers.UpSampling2D(size=(7,7))(inputs)
  base_stuff = base_model(up_sample)
  model = tf.keras.models.Model(inputs = inputs,outputs =base_stuff)
  model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),optimizer='sgd',metrics=['accuracy'])

  return model
```

```
def draw_bounding_boxes_on_image_array(images,boxes,color = [],thickness=1,display_str_list=()):
  img_pil =  PIL.Image.fromarray(images)
  rgbimg = PIL.Image.new('RGBA',img_pil.size)
  rgbimg.paste(img_pil)
  draw_bounding_boxes_on_image(rgbimg,boxes,color,thickness,display_str_list)
  return np.array(rgbimg)
def draw_bounding_boxes_on_image(images,boxes,color =[],thickness=1,display_str_list=()):
  boxes_shape = boxes.shape
  if not boxes_shape :
    return
  if len(boxes_shape) !=2 or boxes_shape[1] !=4:
    raise ValueError('Input must be of size [N,4]')
  for i in range(boxes_shape[0]):
    draw_bounding_box_on_image(images,boxes[i,1],boxes[i,0],boxes[i,3],boxes[i,2],color[i],thickness,display_str_list[i])
def draw_bounding_box_on_image(images,ymin,xmin,ymax,xmax,color ='red',thickness = 1,display_str = None,use_normalized_coordinates=True):
  draw = PIL.ImageDraw.Draw(images)
  img_width,img_height = images.size
  if use_normalized_coordinates:
    (left,right,top,bottom) = (xmin*img_width,xmax*img_width,ymin*img_height,ymax*img_height)
  else:
    (left,right,top,bottom) = (xmin,xmax,ymin,ymax)
  draw.line([(left,top),(left,bottom),(right,bottom),(right,top),(left,top)],width = thickness,fill=color)
```

```
def read_image_tfds(image, label):
    xmin = tf.random.uniform((), 0 , 48, dtype=tf.int32)
    ymin = tf.random.uniform((), 0 , 48, dtype=tf.int32)
    image = tf.reshape(image, (28,28,1,))
    image = tf.image.pad_to_bounding_box(image, ymin, xmin, 75, 75)
    image = tf.cast(image, tf.float32)/255.0
    xmin = tf.cast(xmin, tf.float32)
    ymin = tf.cast(ymin, tf.float32)
   
    xmax = (xmin + 28) / 75
    ymax = (ymin + 28) / 75
    xmin = xmin / 75
    ymin = ymin / 75
    return image, (tf.one_hot(label, 10), [xmin, ymin, xmax, ymax])
```

```
def get_training_dataset():
  with strategy.scope():
    dataset = tfds.load('mnist',as_supervised=True,split ='train',try_gcs=True)
    dataset = dataset.map(read_image_tfds,num_parallel_calls=16)
    dataset = dataset.shuffle(5000,reshuffle_each_iteration=True)
    dataset  = dataset.repeat()
    dataset = dataset.batch(batch_size,drop_remainder = True)
    dataset = dataset.prefetch(-1)
  return dataset
```

```
def feature_extractor(inputs):
  x = tf.keras.layers.Conv2D(16,kernel_size=3,activation='relu',input_shape=(75,75,1))(inputs)
  x = tf.keras.layers.AveragePooling2D((2,2))(x)
  x = tf.keras.layers.Conv2D(32,activation ='relu',kernel_size=3)(x)
  x = tf.keras.layers.AveragePooling2D((2,2))(x)
  x = tf.keras.layers.Conv2D(64,kernel_size=3,activation='relu')(x)
  x = tf.keras.layers.AveragePooling2D((2,2))(x)
  return x 
```

```
def base_model(inputs):
  x = tf.keras.layers.Flatten()(inputs)
  x = tf.keras.layers.Dense(128,activation='relu')(x)
  return x
```

```
def classifier(inputs):
  x = tf.keras.layers.Dense(10,activation='softmax',name = 'classifier')(inputs)
  return x
```

```
def bounding_box_regression(inputs):
  x = tf.keras.layers.Dense(4,activation='softmax',name='regressor')(inputs)
  return x 
```

```
def final_model(inputs):
  features = feature_extractor(inputs)
  base = base_model(features)
  classify = classifier(base)
  regressor = bounding_box_regression(classify)
  model = tf.keras.models.Model(inputs = inputs,outputs=[classify,regressor])
  return model
```

```
def define_and_compile_model(inputs):
  model = final_model(inputs)
  model.compile(optimizer='adam',
                loss={ 'classifier':'categorical_crossentropy',
                'regressor':'mse'
                },metrics = {
                    'classifier':'accuracy','regressor':'mse'
                })
  return model
```

```
def intersection_over_union(pred_box, true_box):
    xmin_pred, ymin_pred, xmax_pred, ymax_pred =  np.split(pred_box, 4, axis = 1)
    xmin_true, ymin_true, xmax_true, ymax_true = np.split(true_box, 4, axis = 1)

    smoothing_factor = 1e-10

    xmin_overlap = np.maximum(xmin_pred, xmin_true)
    xmax_overlap = np.minimum(xmax_pred, xmax_true)
    ymin_overlap = np.maximum(ymin_pred, ymin_true)
    ymax_overlap = np.minimum(ymax_pred, ymax_true)

    pred_box_area = (xmax_pred - xmin_pred) * (ymax_pred - ymin_pred)
    true_box_area = (xmax_true - xmin_true) * (ymax_true - ymin_true)

    overlap_area = np.maximum((xmax_overlap - xmin_overlap), 0)  * np.maximum((ymax_overlap - ymin_overlap), 0)
    union_area = (pred_box_area + true_box_area) - overlap_area
    
    iou = (overlap_area + smoothing_factor) / (union_area + smoothing_factor)

    return iou
```

```
def read_imagetfds(image,box):
  image = tf.cast(image,tf.float32)
  shape = tf.shape(image)
  factor_x = tf.cast(shape[1],tf.float32)
  factor_y = tf.cast(shape[0],tf.float32)
  image = tf.image.resize(image,size=(224,224,))
  image = image/127.5
  image -=1
  box_list = [box[0]/factor_x,box[1]/factor_y,box[2]/factor_x,box[3]/factor_y]
  return image,box_list
```

```
def read_image_with_shape(image,box):
  original_image = image
  image,box_list= read_image_tfds(image,box)
  return original_image,image,box_list
```

```
def read_image_tfds_with_original_box(data):
  image = data['image']
  box = data['bbox']
  shape = tf.shape(image)
  factor_x = tf.cast(shape[1],tf.float32)
  factor_y = tf.cast(shape[0],tf.float32)
  box_list = [box[1]*factor_x,box[0]*factor_y,box[3]*factor_x,box[2]*factor_y]
  return image,box_list
```

```
def dataset_to_numpy_util(dataset,batch_size=0,N=0):
  take_dataset = dataset.shuffle(1024)
  if batch_size>0:
    take_dataset=take_dataset.batch(batch_size)
  if N>0 :
    take_data = take_dataset.take(N)
  if tf.executing_eagerly():
    ds_images,ds_boxes  =[],[]
    for images,boxes in take_data:
      ds_images.append(images.numpy())
      ds_boxes.append(boxes.numpy())
  return (np.array(ds_images),np.array(ds_boxes))
```

```
def dataset_to_numpy_with_original_boxes_util(dataset,batch_size=0,N=0):
  normalized_dataset = dataset.map(read_image_with_shape)
  if batch_size>0:
    normalized_dataset = normalized_dataset.batch(batch_size)
  if N>0:
    normalized_dataset= normalized_dataset.take(N)
  if tf.executing_eagerly():
    ds_original_images,ds_images,ds_boxes =[],[],[]
    for original_images,images,boxes in normalized_dataset:
      ds_original_images.append(original_images.numpy())
      ds_images.append(images.nump())
      ds_boxes.append(boxes.numpy())
  return np.array(ds_original_images),np.array(ds_images),np.array(boxes)
```

```
def get_visualization_validation_dataset():
  dataset,info = tfds.load('caltech_birds2010',try_gcs=True,with_info=True,split='test',data_dir = data_dir)
  visualization_validation_dataset = dataset.map(read_image_tfds_with_original_box,num_parallel_calls=16)
  return visualization_validation_dataset
visualization_validation_dataset = get_visualization_validation_dataset()
(visualization_validation_images,visualization_validation_boxes)= dataset_to_numpy_util(visualization_validation_dataset,N=10)
display_digits_with_boxes(np.array(visualization_validation_images),np.array([]),np.array(visualization_validation_boxes),np.array([]),'validation images and boxes')
```

```
batch_size=64
def get_training_dataset(dataset):
  dataset = dataset.map(read_imagetfds,num_parallel_calls =16)
  dataset = dataset.shuffle(512,reshuffle_each_iteration=True)
  dataset = dataset.repeat()
  dataset = dataset.batch(batch_size)
  dataset = dataset.prefetch(-1)
  return dataset
def get_testing_dataset(dataset):
  dataset= dataset.map(read_imagetfds,num_parallel_calls=16)
  dataset = dataset.batch(batch_size)
  dataset =dataset.repeat()
  return dataset
```

```
training_dataset = get_training_dataset(visualization_training_dataset)
validation_dataset = get_testing_dataset(visualization_validation_dataset)
```

```
def feature_extractor(inputs):
  feature_extractor = tf.keras.applications.MobileNetV2(weights='imagenet',include_top=False,input_shape=(224,224,3))
  x = feature_extractor(inputs)
  return x
```

```
def dense_layers(inputs):
  x = tf.keras.layers.GlobalAveragePooling2D()(inputs)
  x = tf.keras.layers.Flatten()(x)
  x = tf.keras.layers.Dense(1024,activation='relu')(x)
  x = tf.keras.layers.Dense(512,activation = 'relu')(x)
  return x
```

```
def regressor(inputs):
  x  = tf.keras.layers.Dense(4,activation='softmax',name='regressor')(inputs)
  return x
```

```
def final_model(inputs):
  feature_stuff =feature_extractor(inputs)
  dense_stuff = dense_layers(feature_stuff)
  bounding_box_regressor = regressor(dense_stuff)
  model = tf.keras.models.Model(inputs=inputs,outputs=bounding_box_regressor)
  return model
```

```
def build_and_compile_model():
  inputs = tf.keras.layers.Input(shape=(224,224,3))
  model = final_model(inputs)
  model.compile(loss='mse',optimizer=tf.keras.optimizers.SGD(momentum=0.9))
  return model
```

```
model = build_and_compile_model()
model.summary()
```

```
batch_size=32
length_of_training_dataset = len(visualization_training_dataset)
length_of_validation_dataset =len(visualization_validation_dataset)
steps_per_epoch= length_of_training_dataset//batch_size
validation_steps = length_of_validation_dataset//batch_size
if length_of_validation_dataset%batch_size>0:
  validation_steps+=1
```

```
history = model.fit(training_dataset,validation_data= [validation_dataset],steps_per_epoch=steps_per_epoch,validation_steps=validation_steps)
```

```
loss = model.evaluate(validation_dataset,steps=validation_steps)
print('Loss:',loss)
plot_metrics('loss','Bounding Box Loss',ylim=0.2)
def intersection_over_union(pred_boxes,true_boxes):
  xmin_pred,ymin_pred,xmax_pred,ymax_pred = np.split(pred_boxes,4,axis=1)
  xmin_true,ymin_true,xmax_true,ymax_true= np.split(true_boxes,4,axis=1)
  xmin_overlap = np.maximum(xmin_pred,xmin_true)
  xmax_overlap=np.minimum(xmax_pred,xmax_true)
  ymin_overlap= np.maximum(ymin_pred,ymin_true)
  ymax_overlap = np.minimum(ymax_pred,ymax_true)
  true_boxes_area = (xmax_true-xmin_true)*(ymax_true-ymin_true)
  pred_boxes_area= (xmax_pred-xmin_pred)*(ymax_pred-ymin_pred)
  overlap_area = np.maximum((xmax_overlap-xmin_overlap),0)*np.maximum((ymax_overlap-ymin_overlap),0)
  union_area = (true_boxes_area+pred_boxes_area) - overlap_area
  smoothing_factor =1e-10
  iou = (union_area+smoothing_factor)/(overlap_area+smoothing_factor)
  return iou

```

```
original_images,normalized_images,normalized_boxes = dataset_to_numpy_with_original_boxes_util(visualization_validation_dataset,N=500)
predicted_boxes = model.predict(normalized_images,batch_size=32)
iou =intersection_over_union(predicted_boxes,normalized_boxes)
iou_threshold = 0.5

print("Number of predictions where iou > threshold(%s): %s" % (iou_threshold, (iou >= iou_threshold).sum()))
print("Number of predictions where iou < threshold(%s): %s" % (iou_threshold, (iou < iou_threshold).sum()))
n=10
indexes = np.random.choice(len(predicted_boxes),size=n)
iou_to_draw = iou[indexes]
norm_to_draw=original_images[indexes]

display_digits_with_boxes(original_images[indexes], predicted_boxes[indexes], normalized_boxes[indexes], iou[indexes], "True and Predicted values", bboxes_normalized=True)
```
