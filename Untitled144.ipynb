```
def read_image_tfds(image,box):
  image = tf.cast(image,tf.float32)
  shape = tf.shape(image)
  factor_x = tf.cast(shape[0],tf.float32)
  factor_y= tf.cast(shape[1],tf.float32)
  image = tf.image.resize(image,size=(224,224,))
  image = image/127.5
  image -=1
  box_list = [box[0]/factor_x,box[1]/factor_y,box[2]/factor_x,box[3]/factor_y]
  return image,box_list
```

```
def read_image_with_shape(image,box):
  original_image = image
  image,box_list = read_image_tfds(image,box)
  return original_image,image,box_list
```

```
def read_image_tfds_with_original_box(data):
  image = data['image']
  box = data['bbox']
  shape = tf.shape(image)
  factor_x = tf.cast(shape[0],tf.float32)
  factor_y = tf.cast(shape[1],tf.float32)
  box_list = [box[1]/factor_x,box[0]/factor_y,box[3]/factor_x,box[2]/factor_y]
  return image,box_list
```

```
def dataset_to_numpy_util(dataset,batch_size=0,N=0):
  take_dataset = dataset.shuffle(1024)
  if batch_size> 0:
    take_dataset = take_dataset.batch(batch_size)
  if N>0:
    take_dataset = take_dataset.take(N)
  if tf.executing_eagerly():
    ds_images,ds_boxes =[],[]
    for images,boxes in take_dataset:
      ds_images.append(images.numpy())
      ds_boxes.append(boxes.numpy())
  return np.array(ds_images),np.array(ds_boxes)
```

```
def dataset_to_numpy_with_original_boxes_util(dataset,batch_size=0,N=0):
  normalized_dataset = dataset.map(read_image_with_shape)
  if batch_size>0:
    normalized_dataset =normalized_dataset.batch(batch_size)
  if N>0 : 
    normalized_dataset= normalized_dataset.take(N)
  if tf.executing_eagerly():
    ds_original_images,ds_images,ds_boxes =[],[] ,[]
    for original_images,images,boxes in normalized_dataset:
        ds_original_images.append(original_images.numpy())
        ds_images.append(images.numpy())
        ds_boxes.append(boxes.numpy())
  return np.array(ds_original_images),np.array(ds_images),np.array(ds_boxes)
```

```
batch_size=64
def get_training_dataset(dataset):
  dataset = dataset.map(read_image_tfds,num_parallel_calls= 16)
  dataset = dataset.shuffle(512,reshuffle_each_iteration=True)
  dataset = dataset.repeat()
  dataset  = dataset.batch(batch_size)
  datset = dataset.prefetch(-1)
  return dataset
def get_validation_dataset(dataset):
  dataset= dataset.map(read_image_tfds,num_parallel_calls=16)
  dataset = dataset.batch(batch_size)
  dataset = dataset.repeat()
  return dataset
training_dataset = get_training_dataset(visualization_training_dataset)
validation_dataset = get_validation_dataset(visualization_validation_dataset)




```

```
def feature_extractor(inputs):
  mobilenet_v2 = tf.keras.applications.MobileNetV2(weights='imagenet',input_shape=(224,224,3),include_top=False)
  feature_extractor = mobilenet_v2(inputs)
  return feature_extractor
```

```
def dense_layers(inputs):
  x = tf.keras.layers.GlobalAveragePooling2D()(inputs)
  x = tf.keras.layers.Flatten()(x)
  x = tf.keras.layers.Dense(1024,activation='relu')(x)
  x = tf.keras.layers.Dense(512,activation ='relu')(x)
  return x 
```

```
def bounding_box_regression(inputs):
  regression = tf.keras.layers.Dense(4,activation='relu',name='bounding_box')(inputs)
  return regression
  
```

```
def final_model(inputs):
  feature_extract = feature_extractor(inputs)
  dense_layer = dense_layers(feature_extract)
  bounding_box_work = bounding_box_regression(dense_layer)
  model  = tf.keras.Model(inputs = inputs,outputs=bounding_box_work)
  return model
```

```
def build_and_compile_model():
  inputs = tf.keras.layers.Input(shape=(224,224,3))
  model = final_model(inputs)
  model.compile(loss='mse',optimizer = tf.keras.optimizers.SGD(momentum=0.9))
  return model
```

```
model = build_and_compile_model()
model.summary()
```

```
epochs = 1
batch_size=32
length_of_training_dataset = len(visualization_training_dataset)
length_of_validation_dataset = len(visualization_validation_dataset)
steps_per_epoch = length_of_training_dataset//batch_size
validation_steps = length_of_validation_dataset//batch_size
if length_of_validation_dataset%batch_size > 0:
  validation_steps +=1 
```
